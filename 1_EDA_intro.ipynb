## Exploratory Data Analysis

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 
import ast
import seaborn as sns

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df["job_skills"] = df["job_skills"].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

### Filter for US Data Analyst roles

```python
df_DA_US = df[(df["job_title_short"] == "Data Analyst") & (df["job_country"] == "United States")]
df_DA_US
```

```python
df_plot=df_DA_US["job_location"].value_counts().head(10).to_frame()

sns.set_theme(style="ticks")
sns.barplot(data=df_plot, x="count", y="job_location", hue="count", palette="dark:b_r", legend=False)
sns.despine() # per togliere bordo attorno
plt.title("Counts of Job Locations for Data Analyst in the US", fontweight='bold')
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3)
fig.set_size_inches(10, 5)

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(),  startangle=90, autopct="%1.1f%%", labels=["False", "True"])
    ax[i].set_title(title, size=15, fontweight='bold')
    
plt.tight_layout()
plt.show()
```

```python
df_plot=df_DA_US["company_name"].value_counts().head(10).to_frame()

sns.set_theme(style="ticks")
sns.barplot(data=df_plot, x="count", y="company_name", hue="count", palette="dark:b_r", legend=False)
sns.despine() # per togliere bordo attorno
plt.title("Counts of Companies for Data Analyst in the US", fontweight='bold')
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
plt.tight_layout()
```

```python
import pycountry
import pycountry_convert as pc

df_DA_world=df[df["job_title_short"]=="Data Analyst"].copy()

def find_country(country):
        country_name = pycountry.countries.get(name=country)
        
        if country_name is None:
            return "NA"
        
        country_alpha2 = country_name.alpha_2
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        
        continents = {
            'AF': 'Africa',
            'AS': 'Asia',
            'EU': 'Europe',
            'NA': 'North America',
            'SA': 'South America',
            'OC': 'Oceania',
            'AN': 'Antarctica'
        }
        
        return continents.get(continent_code, "NA")

df_DA_world["Continent"]=df_DA_world["job_country"].apply(lambda x: find_country(x) if not pd.isna(x) else "NA")
df_continent=df_DA_world["Continent"].value_counts().to_frame()
```

```python
df_continent
```

```python
sns.set_theme(style="ticks")
sns.barplot(data=df_continent, x="count", y="Continent", hue="count", palette="pastel", legend=False)
sns.despine() 
plt.title("Data Analyst Jobs for Continent", fontweight="bold")
plt.xlabel("Number of Jobs")
ax=plt.gca()

def custom_formatter(x, pos):
    if x < 1000:
        return f'{int(x)}'
    else:
        return f'{int(x/1000)}K'

ax.xaxis.set_major_formatter(plt.FuncFormatter(custom_formatter))
plt.ylabel("")
plt.tight_layout()
plt.show()
```
